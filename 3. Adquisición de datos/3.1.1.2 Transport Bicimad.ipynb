## 3.1.1.2.1. List BiciMAD Stations

```python
URL_STATIONS = "https://openapi.emtmadrid.es/v1/transport/bicimad/stations/"
```

```python
response_stop = requests.request("GET", URL_STATIONS, headers = headers_token)
parsed = json.loads(response_stop.text)["data"]
print(json.dumps(parsed, indent=4))
```

### csv

```python
def print_bicimad_station(data):
    return [data["id"],data["number"],data["name"],data["address"],data["activate"],data["total_bases"]
           ,data["geometry"]["coordinates"][0],data["geometry"]["coordinates"][1]]
```

```python
res = []
for station in parsed:
    res.append(print_bicimad_station(station))
pd.DataFrame(res,columns = ["id","number","name","address","activate","total_bases","longitude","latitude"]).to_csv('bicimad_stations.csv', sep=';', encoding='utf-8',index=False)
```
